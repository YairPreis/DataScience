## Linear Regression

```r
tips <- read.csv("../data/tips.csv")   ## the two points .. is because we want to go back to data dirctory
head(tips)
```

```r
summary(tips)
```

```r
hist(tips$tip)
```

```r
# mod1 <- lm(tips$tip ~ . , tips)  #-> for regretion on all the excist varibles

 mod1 <- lm(tips$tip ~ total_bill+size, tips)  #sicntex: lm(y~x1+x2+x3..., data)

summary(mod1)

```

```r
vif(mod1)  ## vif check if there is high corolation between the varibles.
# higher than 3 it high for strict, or higher than 5 for not strict.

# we see that time have a high corolation (GVIF^(1/2*DF)) so we going back to the regretion and check
# it agin without the varible  "time"
```

```r
Table2(mod1)   #  making the CI (confidens interval) at ramat bithon 5%
```

```r
Table2.forestplot(mod1)
```

```r
pred <- predict(mod1, newdata = tips)  # we can see that the model is not good because the predicted distrebution is very
# differnt from reall distrebution that we saw  at the beggining on this execise. 
hist(pred)


# sintex of predict function: predict(the regretion model, newdata) ->
# we put the same data, not new one, so if our model is good, we should get Frequncy similar to the reall frequncy.
```

```r
scatter.smooth(pred,tips$tip)  #    ?  ? ? ? ?
```
