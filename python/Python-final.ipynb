## Q2. How many students have each course of the English department and the total number of students in the department?

```python
df2 = df1.copy()
df2 = df2[df2.DepartmentID==1]                      # filter the data set by the English department

df3 = df2.groupby(['DepartmentID','CourseId']).agg({'StudentId':[pd.Series.nunique]}) #count by course
df4 = df2.groupby('DepartmentID').agg({'StudentId':[pd.Series.nunique]})              #count by department
df5 = pd.merge(df3, df4,on='DepartmentID')                                            # merge the two tables 
df5.columns                                                                           # check the exact columns names
```

```python
# changing the columns names:
df5.columns = ['Count_by_course','Count_by_Department']
df5
```
