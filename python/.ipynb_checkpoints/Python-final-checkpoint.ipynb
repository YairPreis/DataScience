# Questions
 
 
## Q1. Count the number of students on each department

```python

UnTable.groupby(['DepartmentID','DepartmentName']).agg({'StudentId':['count']})
```

## Q2. How many students have each course of the English department and the total number of students in the department?

```python
UnTable1 = UnTable.copy()

#filttering the table by the English department:

UnTable1 = UnTable1[UnTable.DepartmentID==1]                     

# agrigate the tables:

countbycourse = UnTable1.groupby(['DepartmentName','CourseName']).agg({'StudentId':[pd.Series.nunique]}) #count by course
countbydepartment = UnTable1.groupby('DepartmentName').agg({'StudentId':[pd.Series.nunique]})            #count by department

# merging thr=e two tables above: 

countbycourse.reset_index(inplace=True)
countbydepartment.reset_index(inplace=True)
Q2 = pd.merge(countbycourse,countbydepartment,how='left',on='DepartmentName')         # merge the two tables 

# changing the columns names:
Q2.columns = ['DepartmentName','CourseName','Count_by_course','Count_by_Department']
Q2
```

## Q3. How many small (<22 students) and large (22+ students) classrooms are needed for the Science department?

```python
# filtering only Science department:

UnTable1 = UnTable.copy()
UnTable1 = UnTable1[UnTable1.DepartmentID==2]    

# aggregation the Students by their cources:

UnTable1 = UnTable1.groupby('CourseName').agg({'StudentId':[pd.Series.nunique]})  

# reset the indexces:

UnTable1.reset_index(inplace=True)

# rename the column:

UnTable1.columns = ['CourseName','Nuber_of_student_in_the_class']

# categorize the class size:

UnTable1['Class_size'] = 'Small'
UnTable1.loc[(UnTable1['Nuber_of_student_in_the_class']>22),'Class_size'] = 'Big'
UnTable1
```

## Q4. A feminist student claims that there are more male than female in the College. Justify if the argument is correct

```python
studentsN = students.groupby('Gender').agg({'StudentId':[pd.Series.nunique]})
teachersN = teachers.groupby('Gender').agg({'TeacherId':[pd.Series.nunique]})
un = pd.merge(studentsN,teachersN,how='left',on='Gender')  
un.columns = ['students Number','teachers Number' ]
un
```



## Q5. For which courses the percentage of male/female students is over 70%?

```python
# The number of studet in each class:

UnTable2 = UnTable.copy()
UnTable2 = UnTable2.groupby('CourseName').agg({'StudentId':[pd.Series.nunique]})  
UnTable2.reset_index(inplace=True) 
UnTable2.columns = ['CourseName','Student_number']

# The number of studet in each class by gender:

UnTable3 = UnTable.copy()
UnTable3 = UnTable3.groupby(['CourseName','Student_Gender']).agg({'StudentId':[pd.Series.nunique]})  
UnTable3.reset_index(inplace=True)                            
UnTable3.columns = ['CourseName','Student_Gender','Student_numberG']

# merge the two tables:

UnTable4 = pd.merge(UnTable3,UnTable2,on = 'CourseName')

# calculate the precentage:

UnTable4['Precentage'] = (UnTable4.Student_numberG/UnTable4.Student_number)

# filter over 70% :
UnTable4[UnTable4.Precentage > 0.7] 
```

## Q6. For each department, how many students passed with a grades over 80?

```python
# creating a table with the number of student with degree over 80:

UnTable1 = UnTable.copy()
UnTable1 = UnTable1[UnTable1.degree > 80] 
UnTable1 = UnTable1.groupby('DepartmentName').agg({'StudentId':[pd.Series.nunique]})  
UnTable1.columns = ['number_of_studet_with_degree_over_80']
UnTable1.reset_index(inplace=True)

#creating a table with all the student in the department:

student_per_dep = UnTable.copy()
student_per_dep = student_per_dep.groupby('DepartmentName').agg({'StudentId':[pd.Series.nunique]})
student_per_dep.reset_index(inplace=True) 
student_per_dep.columns = ['DepartmentName','Number_of_student_per_department']

# adding a precentage column: 

UnTable1['Precentage'] = (UnTable1.number_of_studet_with_degree_over_80/student_per_dep.Number_of_student_per_department)
UnTable1
```

## Q7. For each department, how many students passed with a grades under 60?

```python
# creating a table with the number of student with degree under 60:

UnTable1 = UnTable.copy()
UnTable1 = UnTable1[UnTable1.degree < 60] 
UnTable1 = UnTable1.groupby('DepartmentName').agg({'StudentId':[pd.Series.nunique]})  
UnTable1.columns = ['number_of_studet_with_degree_under_60']
UnTable1.reset_index(inplace=True) 

# for calculating the Precentage column i used the student_per_dep varible from the previous question:

UnTable1['Precentage'] = (UnTable1.number_of_studet_with_degree_under_60/student_per_dep.Number_of_student_per_department)
UnTable1
```

## Q8. Rate the teachers by their average student's grades (in descending order).

```python
UnTable1 = UnTable.copy()

# agrigate the mean degree by the Teachers:

UnTable1 = UnTable1.groupby(['TeacherId','Teacher_FirstName','Teacher_LastName']).agg({'degree':['mean']})  

# fixing the coulumns names:

UnTable1.reset_index(inplace=True) 
UnTable1.columns = ['TeacherId','Teacher_FirstName','Teacher_LastName','mean_degree']

# scaling at descending order:

UnTable1.sort_values(ascending = False,by = 'mean_degree')
```

## Q9. Create a dataframe showing the courses, departments they are associated with, the teacher in each course, and the number of students enrolled in the course (for each course, department and teacher show the names).

```python
UnTable1 = UnTable.copy()
UnTable1 = UnTable1.groupby(['DepartmentName','CourseName','Teacher_FirstName','Teacher_LastName']).agg({'StudentId':['count']})  
UnTable1.reset_index(inplace=True) 
UnTable1.columns = ['DepartmentName','CourseName','Teacher_FirstName','Teacher_LastName','Number_of_student_in_class' ]
UnTable1
```

## Q10. Create a dataframe showing the students, the number of courses they take, the average of the grades per class, and their overall average (for each student show the student name).

```python
# creating a table that show how many courses every student takes: 

num_corse_student = UnTable.copy()
num_corse_student = num_corse_student.groupby(['StudentId','Student_FirstName','Student_LastName']).agg({'CourseName':['count']})  
num_corse_student.reset_index(inplace=True) 
num_corse_student.columns = ['StudentId','Student_FirstName','Student_LastName','Number_of_courses']



# Creating a table that show the mean degree for any studet by  Department:

mean_Department = UnTable.copy()
mean_Department = mean_Department.groupby(['StudentId','DepartmentName']).agg({'degree':['mean']}) 
mean_Department.reset_index(inplace=True) 
mean_Department.columns = ['StudentId','DepartmentName','mean_by_Department']

# Creating a table that show the General mean for any student:

general_mean = UnTable.copy()
general_mean = general_mean.groupby('StudentId').agg({'degree':['mean']})  
general_mean.reset_index(inplace=True) 
general_mean.columns = ['StudentId','general_mean_degree']

```

```python
# merging the tabels above:

degree_means = pd.merge(mean_Department,general_mean,on = 'StudentId') # merge the two means tabels

pd.merge(num_corse_student,degree_means) # merge the united mean table with the num_corse_student table
```
